```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
```

```python
from tqdm import tqdm

base_url = f"https://dungeonsanddragons.fandom.com"
url = f"{base_url}/it/wiki/Oggetti_Homebrew"
data = []
index_page = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})
index_soup = BeautifulSoup(index_page.text, 'html.parser')
magic_items = index_soup.find("div", {"class": "page-content"}).find_all("ul")[-1]
for magic_item in tqdm(magic_items):
    name = magic_item.text.strip()
    if not name:
        continue
    item_page = requests.get(f"{base_url}{magic_item.find('a')['href']}")
    item_soup = BeautifulSoup(item_page.text, 'html.parser')
    rarity = item_soup.find("div", {"class": "mw-parser-output"}).find_all("p")[0].text.strip()
    description = item_soup.find("div", {"class": "mw-parser-output"}).find_all("p")[1].text.strip()
    data += [{
        "name": name,
        "rarity": rarity,
        "description": description,
    }]
data = pd.DataFrame(data)
data.to_csv("data/oggetti_magici.csv", index=False)
data
```
